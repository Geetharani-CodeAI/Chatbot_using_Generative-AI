#### Helper Function 
Temperature

For coding / debugging / Very structured → temperature = 0

For chat / creative writing → 0.7 – 1

For factual answers → 0 – 0.3

```python
import os
import google.generativeai as genai

genai.configure(api_key = "API_Key")

model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
def get_completion(prompt):
    chat= model.start_chat(history = [{"role": "user", "parts" : prompt}])
    response = chat.send_message(prompt, generation_config={"temperature":0})
  
    return response.text

print(get_completion("How to improve focus"))
```

#### Prompt the model and get a completion

```python
print(get_completion("What is the capital of France?"))
```

#### Tokens

```python
print(get_completion("Take the letters in lollipop and reverse it"))
```

```python
def get_completion_and_token_count(prompt):
    chat = model.start_chat(history = [
    {'role':'user', 
     'parts': ["""You are an assistant who response in the style of Dr.Paul"""]},
     #{'role':'user', 
     #'parts': ["""Your response must be in one sentence long"""]}
])
    
    response = chat.send_message(prompt, generation_config = {"temperature":0, "max_output_tokens" : 500})
    content = response.text
    token = response.usage_metadata
    #print("Response:", content)
    #print("Token Usage:", tokens)

    return content, token
```
