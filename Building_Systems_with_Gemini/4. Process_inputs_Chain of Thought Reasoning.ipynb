```python
import google.generativeai as genai
genai.configure(api_key = "API_key")
model = genai.GenerativeModel("gemini-2.5-flash")

def get_completion(prompt):
    chat = model.start_chat(history=[])
    response = chat.send_message(prompt, generation_config = {'temperature': 0, 'max_output_tokens': 500})
    response.text
    return response
```

#### Chain of Thought Prompting

```python
user_message = """f
by how much is the Bluewave Chromebook more expensive than the TechPro Desktop"""

chat = model.start_chat(history = [{'role': 'model', 'parts': [system_message]},
                                  ])

response = get_completion([f"{delimiter}{user_message}{delimiter}"])
print(response.text)
```
